```python
import os

# Function to list files in a directory
def list_files(dir_path):
    return os.listdir(dir_path)

# List files in each directory
print("Training NORMAL files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/train/NORMAL'))
print("Training PNEUMONIA files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/train/PNEUMONIA'))
print("Validation NORMAL files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/val/NORMAL'))
print("Validation PNEUMONIA files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/val/PNEUMONIA'))
print("Test NORMAL files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/test/NORMAL'))
print("Test PNEUMONIA files:", list_files('/content/drive/MyDrive/Kaggle_dataset/chest_xray/test/PNEUMONIA'))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define directories
train_normal_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/train/NORMAL'
train_pneumonia_dir = '/content/drive/MyDrive/Kaggle_dataset/chet_xray/train/PNEUMONIA'
val_normal_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/val/NORMAL'
val_pneumonia_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/val/PNEUMONIA'
test_normal_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/test/NORMAL'
test_pneumonia_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/test/PNEUMONIA'

# Create ImageDataGenerators
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load images from directories
train_generator = train_datagen.flow_from_directory(
    directory='/content/drive/MyDrive/Kaggle_dataset/chest_xray/train',
    classes=['NORMAL', 'PNEUMONIA'],
    target_size=(150, 150),  # Adjust size as needed
    batch_size=32,
    class_mode='binary'  # Use 'categorical' if you have more than two classes
)

val_generator = val_datagen.flow_from_directory(
    directory='/content/drive/MyDrive/Kaggle_dataset/chest_xray/val',
    classes=['NORMAL', 'PNEUMONIA'],
    target_size=(150, 150),  # Adjust size as needed
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    directory='/content/drive/MyDrive/Kaggle_dataset/chest_xray/test',
    classes=['NORMAL', 'PNEUMONIA'],
    target_size=(150, 150),  # Adjust size as needed
    batch_size=32,
    class_mode='binary'
)
```

```python
import matplotlib.pyplot as plt

# Get a batch of images and labels
images, labels = next(train_generator)

# Plot the first image and its label
plt.imshow(images[0])
plt.title(f"Label: {train_generator.class_indices}")
plt.axis('off')
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')  # Use 'softmax' if more than two classes
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(
    train_generator,
    epochs=10,  # Adjust number of epochs as needed
    validation_data=val_generator
)
```


The model training results show that the model achieved high accuracy on the training set but lower performance on the validation set, with test accuracy around 77%. This suggests a few things might be happening:

Overfitting: The model might be overfitting to the training data, especially if it performs significantly better on training data compared to validation and test data.

Model Complexity: The model might be too complex for the amount of training data available, leading to good training performance but poor generalization.

Data Imbalance: If there is an imbalance between the number of images in the NORMAL and PNEUMONIA classes, the model might not generalize well.

Learning Rate and Epochs: The learning rate might need adjustment, or the number of epochs might be too high.

```python
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test Accuracy: {test_accuracy:.2f}")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator,
    callbacks=[early_stopping]
)
```

```python
from tensorflow.keras.layers import Dropout

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.5),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Dropout(0.5),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001)

model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.utils import class_weight
import numpy as np # Import numpy

class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes
)

class_weights = dict(enumerate(class_weights))
```

```python
history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator,
    class_weight=class_weights
)
```

```python
# Import necessary libraries
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set the directory for the test set
test_dir = '/content/drive/MyDrive/Kaggle_dataset/chest_xray/test'

# Define the ImageDataGenerator for the test set without any data augmentation
test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Load the test data
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),  # Use the same target size as used in training
    batch_size=32,           # Adjust the batch size if necessary
    class_mode='binary',     # Use 'binary' since it's a binary classification problem
    shuffle=False            # Do not shuffle the test set
)
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(test_generator)

print(f"Test Accuracy: {test_accuracy:.2f}")
print(f"Test Loss: {test_loss:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Predict the test set labels
test_generator.reset()  # Ensure the generator starts from the beginning
predictions = model.predict(test_generator)
predicted_classes = np.where(predictions > 0.5, 1, 0)

# Get the true labels from the test generator
true_classes = test_generator.classes

# Compute the confusion matrix
conf_matrix = confusion_matrix(true_classes, predicted_classes)

# Plot the confusion matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=test_generator.class_indices.keys(), yticklabels=test_generator.class_indices.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Print the classification report
print(classification_report(true_classes, predicted_classes, target_names=test_generator.class_indices.keys()))
```
